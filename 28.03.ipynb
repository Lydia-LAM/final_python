```python
plt.figure(figsize=(10, 6))
plt.scatter(df['sqft_living'], df['price'], alpha=0.5, color='skyblue')
plt.title('Зависимость стоимости недвижимости от площади квартиры')
plt.xlabel('Площадь квартиры, кв. футов')
plt.ylabel('Стоимость недвижимости, $')
plt.grid(True)
```

Выводы и наблюдения: Как показано на диаграмме, существует положительная корреляция между площадью квартиры и её стоимостью. Обычно чем больше площадь квартиры, тем выше её стоимость.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=df, palette='pastel')
plt.title('Зависимость стоимости недвижимости от числа спален')
plt.xlabel('Число спален')
plt.ylabel('Стоимость недвижимости, $')
```

Выводы и наблюдения: На этой диаграмме мы видим, как стоимость недвижимости изменяется в зависимости от числа спален. Обычно квартиры с большим числом спален имеют более высокую стоимость.

```python
plt.figure(figsize=(8, 6))
plt.hist(df[df['condition'] == 5]['price'], alpha=0.5, color='blue', label='Состояние 5', bins=20)
plt.hist(df[df['condition'] < 5]['price'], alpha=0.5, color='red', label='Состояние < 5', bins=20)
plt.title('Распределение стоимости недвижимости по состоянию домов')
plt.xlabel('Стоимость недвижимости, $')
plt.ylabel('Частота')
plt.legend();
```

Выводы и наблюдения: Эта гистограмма показывает распределение стоимости недвижимости для домов в зависимости от их состояния. Мы видим, что дома с более высоким состоянием имеют более высокие цены.

```python
sns.pairplot(df[['price', 'sqft_living', 'bedrooms']])
```

Выводы и наблюдения: Мы можем видеть, что общая площадь дома и количество спален имеют сильное влияние на его цену. Большая общая площадь и количество спален чаще всего связаны с более высокой ценой.

```python
avg_price_year = df.groupby('yr_built')['price'].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_price_year.index, avg_price_year.values, marker='o', linestyle='-')
plt.title('Средняя стоимость домов в зависимости от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Средняя цена дома, $')
plt.grid(True)
```

На этой линейной диаграмме мы можем увидеть, как меняется средняя стоимость домов в зависимости от года их постройки. Это позволяет нам оценить динамику изменения цен на дома в течение времени и возможно выявить какие-то тенденции или закономерности.
